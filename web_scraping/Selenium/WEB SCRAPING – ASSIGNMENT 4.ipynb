### 1. Scrape the details of most viewed videos on YouTube from Wikipedia. Url = https://en.wikipedia.org/wiki/List_of_most-viewed_YouTube_videos You need to find following details: 

A) Rank

B) Name

C) Artist

D) Upload date

E) Views

```python
import re

import pandas as pd
from selenium.common import TimeoutException
from selenium.webdriver import Keys, ActionChains
from selenium.webdriver.common.by import By
from selenium import webdriver
from selenium.webdriver.support import expected_conditions
from selenium.webdriver.support.wait import WebDriverWait

driver=webdriver.Edge()
driver.maximize_window()
driver.get("https://en.wikipedia.org/wiki/List_of_most-viewed_YouTube_videos")
actions = ActionChains(driver)

view_table_XPATH='//table[@class="sortable wikitable sticky-header static-row-numbers sort-under col3center col4right jquery-tablesorter"]/tbody'
while 1:
    try:
        view_table=WebDriverWait(driver,10).until(expected_conditions.presence_of_element_located((By.XPATH,view_table_XPATH)))
        break
    except TimeoutException:
        print("Timed out..trying again")

rows=view_table.find_elements(By.TAG_NAME,'tr')
video_df=pd.DataFrame(columns=['Rank','Name','Artist','Date','Views (billions)'])

for rank, row in enumerate(rows):
    name=row.find_elements(By.TAG_NAME,'td')[0].text.strip()
    artist=row.find_elements(By.TAG_NAME,'td')[1].text.strip()
    views=row.find_elements(By.TAG_NAME,'td')[2].text.strip()
    date=row.find_elements(By.TAG_NAME,'td')[3].text.strip()
    
    video_df.loc[rank]=[rank+1,name,artist,date,views]
driver.quit()
video_df
```

### 4. Scrape the details of trending repositories on Github.com.

Url = https://github.com/

You have to find the following details:

A) Repository title

B) Repository description

C) Contributors count

D) Language used

Note: - From the home page you have to click on the trending option from Explore menu through code.

```python
import re
import pandas as pd
from selenium.webdriver import Keys, ActionChains
from selenium.webdriver.common.by import By
from selenium import webdriver
from selenium.webdriver.support import expected_conditions
from selenium.webdriver.support.wait import WebDriverWait
from selenium.common.exceptions import TimeoutException

driver=webdriver.Edge()
driver.maximize_window()
driver.get("https://github.com/")

oen_source_XPATH='/html[1]/body[1]/div[1]/div[1]/header[1]/div[1]/div[2]/div[1]/nav[1]/ul[1]/li[3]/button[1]'
open_source_dropdown=WebDriverWait(driver,10).until(expected_conditions.presence_of_element_located((By.XPATH, oen_source_XPATH)))
ActionChains(driver).move_to_element(to_element=open_source_dropdown).perform()

trending_XPATH='/html[1]/body[1]/div[1]/div[1]/header[1]/div[1]/div[2]/div[1]/nav[1]/ul[1]/li[3]/div[1]/div[3]/ul[1]/li[2]/a[1]'
trending=WebDriverWait(driver,10).until(expected_conditions.presence_of_element_located((By.XPATH, trending_XPATH)))
ActionChains(driver).move_to_element(to_element=trending).click(on_element=trending).perform()

trending_repos_XPATH='//article[@class="Box-row"]'
trending_repos=WebDriverWait(driver,10).until(expected_conditions.presence_of_all_elements_located((By.XPATH, trending_repos_XPATH)))

parent_window=driver.current_window_handle

trending_repos_df=pd.DataFrame(columns=['Title', 'Description', 'Contributors count', 'Language'])
for i,repo in enumerate(trending_repos):
    repo_link=repo.find_element(By.XPATH, './/a[@class="Link"]')
    ActionChains(driver).move_to_element(to_element=repo_link).key_down(Keys.CONTROL).click(on_element=repo_link).key_up(Keys.CONTROL).perform()
    driver.switch_to.window(driver.window_handles[-1])
    
    author=WebDriverWait(driver,10).until(expected_conditions.presence_of_element_located((By.XPATH, '//span[@class="author flex-self-stretch"]/a'))).text.strip()
    name=WebDriverWait(driver,10).until(expected_conditions.presence_of_element_located((By.XPATH, '//strong[@class="mr-2 flex-self-stretch"]/a'))).text.strip()
    title=f'{author}/{name}'
    description_XPATH='/html[1]/body[1]/div[1]/div[4]/div[1]/main[1]/turbo-frame[1]/div[1]/div[1]/div[1]/div[2]/div[2]/div[1]/div[1]/div[1]/div[1]/p[1]'
    try:
        description=WebDriverWait(driver,10).until(expected_conditions.presence_of_element_located((By.XPATH, description_XPATH))).text.strip()
    except TimeoutException:
        description='-'
    contributor_count_XPATH='//a[@class="Link--primary no-underline Link d-flex flex-items-center"]'
    try:
        contributor_count=WebDriverWait(driver,10).until(expected_conditions.presence_of_all_elements_located((By.XPATH, contributor_count_XPATH)))[-1].text.strip()
        contributor_count_num=re.compile(pattern='\d+').search(string=contributor_count)
        if contributor_count_num is not None: contributor_count=contributor_count_num.group()
    except TimeoutException:
        contributor_count='-'
    language_XPATH='//span[@class="color-fg-default text-bold mr-1"]'
    languages=''
    try:
        for language in WebDriverWait(driver,10).until(expected_conditions.presence_of_all_elements_located((By.XPATH, language_XPATH))):
            languages+=language.text.strip()+','
    except TimeoutException:
        languages='-'
    trending_repos_df.loc[i]=(title, description, contributor_count, languages[:-1])
    
    driver.close()
    driver.switch_to.window(parent_window)
driver.quit()
trending_repos_df
```

### 8. Details of Datasets from UCI machine learning repositories.

Url = https://archive.ics.uci.edu/ You have to find the following details:

A) Dataset name

B) Data type

C) Task

D) Attribute type

E) No of instances

F) No of attribute 

G) Year

Note: - from the home page you have to go to the Show All Dataset page through code.

```python
import pandas as pd
from selenium.webdriver.common.by import By
from selenium import webdriver
from selenium.webdriver.support import expected_conditions
from selenium.webdriver.support.wait import WebDriverWait
from selenium.webdriver.common.action_chains import ActionChains

driver = webdriver.Chrome()
driver.maximize_window()
driver.get("https://archive.ics.uci.edu/")

all_datasets_XPATH = '/html[1]/body[1]/div[1]/div[1]/div[1]/header[1]/nav[1]/ul[1]/li[1]/a[1]'
all_datasets = WebDriverWait(driver, 10).until(expected_conditions.presence_of_element_located((By.XPATH, all_datasets_XPATH)))
ActionChains(driver).move_to_element(to_element=all_datasets).click(on_element=all_datasets).perform()

datasets_df = pd.DataFrame(columns=['Name', 'Type', 'Task', 'Attribute type', 'No of instances', 'No of attributes', 'Year'])
hrefs=[]
while True:  
    try:
        page_hrefs=WebDriverWait(driver, 10).until(expected_conditions.presence_of_all_elements_located((By.XPATH, '//a[@class="link-hover link text-xl font-semibold"]')))
    except Exception as e:
        print('Page has no dataset link; exiting..')
        break
    for href in page_hrefs:
        hrefs.append(href.get_attribute('href'))
    try:
        next_button_XPATH='/html[1]/body[1]/div[1]/div[1]/div[1]/main[1]/div[1]/div[2]/div[3]/div[1]/button[2]'
        next_button=WebDriverWait(driver,10).until(expected_conditions.presence_of_element_located((By.XPATH, next_button_XPATH)))
        if next_button.is_enabled():
            driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")
            ActionChains(driver).move_to_element(to_element=next_button).click(on_element=next_button).pause(5).perform()
        else:
            print('All pages exhausted')
            break
    except Exception as e:
        print('Exception occured, aborting....')
        break

for i,href in enumerate(hrefs):
    driver.get(href)
    name_XPATH = '/html[1]/body[1]/div[1]/div[1]/div[1]/main[1]/div[1]/div[1]/div[1]/div[1]/div[2]/div[1]/h1[1]'
    try:
        name = WebDriverWait(driver, 10).until(expected_conditions.presence_of_element_located((By.XPATH, name_XPATH))).text.strip()
    except Exception as e:
        name='-'
    type_XPATH = '/html[1]/body[1]/div[1]/div[1]/div[1]/main[1]/div[1]/div[1]/div[1]/div[2]/div[2]/div[1]/p[1]'
    try:
        type = WebDriverWait(driver, 10).until(expected_conditions.presence_of_element_located((By.XPATH, type_XPATH))).text.strip()
    except Exception as e:
        type='-'
    task_XPATH = '/html[1]/body[1]/div[1]/div[1]/div[1]/main[1]/div[1]/div[1]/div[1]/div[2]/div[2]/div[3]/p[1]'
    try:
        task = WebDriverWait(driver, 10).until(expected_conditions.presence_of_element_located((By.XPATH, task_XPATH))).text.strip()
    except Exception as e:
        task='-'
    attribute_type_XPATH = '/html[1]/body[1]/div[1]/div[1]/div[1]/main[1]/div[1]/div[1]/div[1]/div[2]/div[2]/div[4]/p[1]'
    try:
        attribute_type = WebDriverWait(driver, 10).until(expected_conditions.presence_of_element_located((By.XPATH, attribute_type_XPATH))).text.strip()
    except Exception as e:
        attribute_type='-'
    no_of_instances_XPATH = '/html[1]/body[1]/div[1]/div[1]/div[1]/main[1]/div[1]/div[1]/div[1]/div[2]/div[2]/div[5]/p[1]'
    try:
        no_of_instances = WebDriverWait(driver, 10).until(expected_conditions.presence_of_element_located((By.XPATH, no_of_instances_XPATH))).text.strip()
    except Exception as e:
        no_of_instances='-'
    no_of_atributes_XPATH = '/html[1]/body[1]/div[1]/div[1]/div[1]/main[1]/div[1]/div[1]/div[1]/div[2]/div[2]/div[6]/p[1]'
    try:
        no_of_attributes = WebDriverWait(driver, 10).until(expected_conditions.presence_of_element_located((By.XPATH, no_of_atributes_XPATH))).text.strip()
    except Exception as e:
        no_of_attributes='-'
    year_XPATH = '/html[1]/body[1]/div[1]/div[1]/div[1]/main[1]/div[1]/div[1]/div[1]/div[1]/div[2]/h2[1]'
    try:
        year = WebDriverWait(driver, 10).until(expected_conditions.presence_of_element_located((By.XPATH, year_XPATH))).text.strip()
    except Exception as e:
        year='-'
    datasets_df.loc[i] = (name, type, task, attribute_type, no_of_instances, no_of_attributes, year)
driver.quit()
datasets_df
```
